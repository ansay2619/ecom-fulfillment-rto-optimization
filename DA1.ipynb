```python
import pandas as pd
import random

print("🚀 Step 1: Generating raw e-commerce supply chain dataset...")
random.seed(42) # Keeps the dataset consistent

brands = ["ZariBrands", "StepWell", "ElectroTech", "KicksCo", "DesiWeaves"]
messy_products = {
    "ZariBrands": ["Fancy Designer Georgette Pink Saree", "Women's Silk Saree Regular", "Traditional Embroidered Kurti M"],
    "StepWell": ["Formal Leather Shoes Men", "Casual Daily Wear Sandals", "Premium Running Sneakers"],
    "ElectroTech": ["Wireless Bluetooth Earphones Pro", "Fast Charging Type-C Cable", "Smart Watch Series 5"],
    "KicksCo": ["Sporty Canvas Shoes Blue", "Leather Loafers Premium", "Unisex Running Joggers"],
    "DesiWeaves": ["Cotton Palazzo Pants Casual", "Designer Chikankari Kurta Luxury", "Festive Wear Lehenga Set"]
}

cities = ["Patna", "Ranchi", "Delhi", "Bengaluru", "Mumbai", "Guwahati", "Lucknow", "Indore"]
warehouses = ["WH-101 (Mumb)", "WH-102 (Delh)", "WH-103 (Blru)", "WH-104 (Kolkat)"]

data = []
for order_id in range(10001, 11001):
    brand = random.choice(brands)
    raw_title = random.choice(messy_products[brand])
    cust_city = random.choice(cities)
    warehouse = random.choice(warehouses)
    payment = random.choices(["COD", "Prepaid"], weights=[0.65, 0.35])[0] 
    val = round(random.uniform(300, 4500), 2)
    
    # Inherent Risk Triggers: Patna/Ranchi/Guwahati + COD have built-in high return rates
    if (cust_city in ["Patna", "Ranchi", "Guwahati"] and payment == "COD" and random.random() < 0.48):
        status = "Returned"
    elif (brand == "StepWell" and random.random() < 0.30): 
        status = "Returned"
    else:
        status = "Delivered"
        
    # Warehouse Bottleneck: WH-104 (Kolkata) systematically breaches shipping SLAs
    if warehouse == "WH-104 (Kolkat)":
        hours_to_dispatch = random.randint(12, 72) 
    else:
        hours_to_dispatch = random.randint(4, 30)

    data.append([order_id, warehouse, cust_city, payment, brand, raw_title, val, hours_to_dispatch, status])

df_raw = pd.DataFrame(data, columns=["Order_ID", "Warehouse", "Customer_City", "Payment_Method", "Brand_Name", "Raw_Product_Title", "Order_Value_INR", "Hours_to_Dispatch", "Order_Status"])

print("🧹 Step 2: Running Python text-cleaning and product categorization pipeline...")

def categorize_product(title):
    title_lower = title.lower()
    if 'saree' in title_lower or 'kurti' in title_lower or 'palazzo' in title_lower or 'kurta' in title_lower or 'lehenga' in title_lower:
        return "Ethnic Wear"
    elif 'shoes' in title_lower or 'sandals' in title_lower or 'sneakers' in title_lower or 'loafers' in title_lower or 'joggers' in title_lower:
        return "Footwear"
    elif 'earphones' in title_lower or 'cable' in title_lower or 'watch' in title_lower:
        return "Electronics"
    else:
        return "General Merchandise"

# The Core Analyst Move: Creating a clean, normalized column
df_raw["Product_Category"] = df_raw["Raw_Product_Title"].apply(categorize_product)
df_clean = df_raw.drop(columns=["Raw_Product_Title"])
df_clean = df_clean[["Order_ID", "Warehouse", "Customer_City", "Payment_Method", "Brand_Name", "Product_Category", "Order_Value_INR", "Hours_to_Dispatch", "Order_Status"]]

# Save the dataset file locally to the Colab instance
df_clean.to_csv("clean_ecommerce_fulfillment_data.csv", index=False)
print("🎉 Success! Your clean dataset file is ready: 'clean_ecommerce_fulfillment_data.csv'")

# Preview the first few rows of your data
print("\n👀 Data Preview:")
print(df_clean.head())
```
